```python
class Circle:
    def __init__(self,pi,radius,diameter):
        self.pi = pi
        self.radius = radius
        self.diameter = diameter

    def area(self):
        options = ['diameter','radius']
        option = str(input("Do you the diameter or radius.....?")).strip().lower()
        if option not in options:
            while option not in options:
                option = str(input("Please try again and only enter diameter or radius....!!!"))
        elif option == "diameter" :
            diameter = float(input("What is the value of the diameter?"))
            radius = diameter/2
            area = self.pi * radius ** 2
            print("The area of the Circle is: ",area)
        elif option == "radius" :
            radius = float(input("What is the value of the radius?"))
            area = self.pi * radius ** 2
            print("The area of the Circle is: ",area)

    def perimeter(self):
        options = ['diameter','radius']
        option = str(input("Do you the diameter or radius.....?")).strip().lower()
        if option not in options:
            while option not in options:
                option = str(input("Please try again and only enter diameter or radius....!!!"))
        elif option == "diameter" :
            diameter = float(input("What is the value of the diameter?"))
            radius = diameter/2
            perimiter = 2 * self.pi * radius
            print("The perimiter of the Circle is: ",perimiter)
        elif option == "radius" :
            radius = float(input("What is the value of the radius?"))
            perimiter = 2 * self.pi * radius
            print("The perimiter of the Circle is: ",perimiter)



def main_menu():
    pi = 3.141592653589793238462643383279502884197
    radius = 0
    diameter = 0 
    circle = Circle(pi,radius,diameter)
    while True:
        print("\n====== Main Menu ====== ")
        print("1.Calculate Area")
        print("2.Calculate Perimeter")
        print("3.Exit")

        choice = input("Enter your choice (1-3): ")

        if choice == '1':
            circle.area()
        elif choice == '2':
            circle.perimeter()
        elif choice == '3':
            print("Your are exsiting the program.\Thank you for using our program have a GoodDAY")
            break
        else:
            print("Invalid choice. Please Try again.")

main_menu()
```
